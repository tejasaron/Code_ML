```python
# Utility Function To Align, Sort, Fill, & Drop NA's

def align_forward_fill(dfs, join='outer'):
  df = pd.concat(dfs, axis = 1, join = join) # aligns the frames
  df.sort_index(inplace = True) # sorts the frames
  df = df.fillna(method = 'ffill').dropna # first its fills values & then drops values.
  return df
```

```python
# Moving Features Utility

def add_moving_feature(df, col='close', windows=[3,5,10,20]):
  for w in windows:
    df[f'return_{w}'] = df[col].pct_change(w) # calculates percentage change
    df[f'rolling_mean_{w}'] = df[col].rolling(window = w).mean()
    df[f'rolling_std_{w}'] = df[col].rolling(window = w).std()
  return df
```
